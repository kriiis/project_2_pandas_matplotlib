Before we create a visualisation, we always need to prepare our dataframe: 

```python
p_df = (
orders_compl
    # create two new columns in our dataframe; year and month
    .assign(year = orders['created_date'].dt.strftime('%Y'), 
           month = orders['created_date'].dt.strftime('%b'))
    # get the total paid by each year and month
    .groupby(['year','month'])
    .agg({'total_paid':'sum', 'id_order':'count'})
     #reset the index to use the column vector in our plot
    .reset_index()
)
p_df
```

```python
ol_df = (
orderlines
    # create two new columns in our dataframe; year and month
    .assign(year = orderlines['date'].dt.strftime('%Y'), 
           month = orderlines['date'].dt.strftime('%b'))
    .merge(orders, how='left', on ='id_order')
    .query('state == "Completed"')
    .groupby(['year','month'])
    .agg({'total_price':'sum'})
     #reset the index to use the column vector in our plot
    .reset_index()
    #.value_counts('id_order')
)
ol_df
```

```python
#see the divergence between total_paid and total_price for each month in each year as percentage
test=p_df.merge(ol_df)
test['percentage']=(test.total_price *100)/test.total_paid
test
```

```python
# Create a list with the months sorted from January to December
# It is going to help us sort our columns appropriately
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 
          'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']

# define de size of the plot: 12 with be the width, and 8 the height
plt.subplots(figsize=(12, 8))


p = sns.barplot(
    data=p_df, # dataframe we will use to look for the columns information
    x='month', # x values on our chart
    y='total_paid', # y values on our chart
    hue='year', # group color in our chart
    order=months # order x columns
)
# add a horizontal reference line into the chart to know where the mean value will be
p.axhline(
    p_df['total_paid'].mean(), # value where you want to add your reference line
    linestyle='--', # type of line you would like to see on the plot. You can find more informaiton on its documentation
    color='gray'
)

plt.xlabel('Month') # x label name
plt.ylabel('Total paid') # y label name
plt.title('Total paid evolution through 2017 and 2018') # title
plt.ticklabel_format(style='plain', axis='y') # remove scientific notation for the y axis
plt.show() # show the plot
```

## Looking for a correlation between shipping cost, total price and quantity of products by each order 

Is there a correlation between the number of products acquired and the shipping cost? 

```python
p_df = (
orderlines
    .groupby(['id_order'])
    .agg({'total_price':'sum', 
         'product_quantity':'sum'})
    .reset_index()
    .merge(orders, how='left', left_on='id_order', right_on='id_order')
    .query('state == "Completed"')
    # lambda function inside assign help us to call a variable create on the pandas pipeline
    # further explanation below
    .assign(ship_cost = lambda x: x['total_paid'] - x['total_price'])
    # exclude orders with no shippping cost
    .query('ship_cost != 0')
    .groupby('id_order')
    .agg({'ship_cost':'sum',
         'total_price':'sum', 
         'product_quantity':'sum'})
    .sort_values('product_quantity')
)
```

```python
# Example how to understand a lambda function inside assign

# create a dummy dataframe
temp = pd.DataFrame({
    'x':[1,2,3],
    'y':[2,3,4]
})

# example how it is applied
print('Example 1:')
print(
temp
    .assign(x_y = temp['x'] + temp['y'])
    # the only way to call the variable x_y in the same data pipeline is to use 
    # a lambda function, as the variable x_y do not exist on the original dataframe
    .assign(x__x_y = lambda x: x['x'] + x['x_y'])
)

# this process will be exaclty as doing
print('\nExample 2:')
temp = temp.assign(x_y = temp['x'] + temp['y'])
# in this example I can call the variable as temp['x_y'] as I overwrote the object temp in the line above
print(temp.assign(x__x_y = temp['x'] + temp['x_y']))
```

```python
plt.subplots(figsize=(10, 8))

sns.scatterplot(data=p_df, x='ship_cost', y='total_price', 
                hue='product_quantity', 
                size='product_quantity', # define the size of each point based on the variable product quantity
                sizes=(20, 200), # range of size of the size variable
                palette='ch:start=.2,rot=-.3') # define manually the range of colors

plt.xlabel('Shipping cost in Euros')
plt.ylabel('Total price in Euros')
plt.title('Relation between shipping cost, total price and quantity of products by order id')
plt.legend(title = 'Product quantity') # Change the title name of the legend
plt.xlim(-7, 100)
sns.despine()
plt.show()
```

a. Show the evolution of the average discount (in %) by each month
    

```python
prices_merge = (
    orderlines
    .merge(orders, how='left', on='id_order')
    .query('state == "Completed"')
    .merge(products, on='sku', how='left')
    #as some prices are in column max_price_ol to replace missing values in prices
    #assign those max_price_ol to the prices column where there are missing values
    .assign(price = lambda x: np.where(np.isnan(x['price']), x['max_price_ol'], x['price']))
    .assign(discount = lambda x: x['price'] - x['unit_price'],
        discount_percentage = lambda x: (x['discount'] * 100) / x['price'],
        year_month = lambda x: x['date'].dt.to_period('M'))
)
prices_merge.head(2)
```

```python
start_47 = '2017-11-20'
end_47 = '2017-11-26'

mask = (prices_merge['date'] > start_47) & (prices_merge['date'] <= end_47)
week_47 = prices_merge.loc[mask]
week_47 = week_47.agg({'discount':'mean', 'discount_percentage':'mean'})
week_47
```

```python
discount_evol = (
    prices_merge
    .groupby(['year_month'])
    .agg({'discount_percentage':'mean', 'discount':'mean'})
    .reset_index()
)
discount_evol.head(2)
```

```python
plt.subplots(figsize=(12, 8))

plt.plot_date(data=discount_evol, x='year_month', y='discount_percentage', linestyle='-')

plt.ylim(0, discount_evol['discount_percentage'].max()*1.2)
plt.title('Evolution of discounts (%) overtime')
plt.ylabel('Average discount (%)')
sns.despine()
plt.show()
```

```python
def start_pipeline(data): 
    return data.copy()

def filter_outliers(data_o, data_ol):
    data_ol['total_price'] = data_ol['unit_price'] * data_ol['product_quantity']
    price_info = (
    data_ol
        .groupby('id_order')
        .agg({'total_price':'sum'})
        .merge(data_o, how='inner', on='id_order')
        .copy()
    )
    price_info= (
    price_info
        .assign(price_difference = round(price_info['total_paid'] - price_info['total_price'], 2)) 
    )
    ids = (
    price_info
        .query('price_difference >= 0 & price_difference <= 105')
        ['id_order'].tolist()
    )
    return ids

def remove_outliers(data, id_list):
    return data.query('id_order == @id_list')
```

```python
orders_clean = (
orders
    .pipe(start_pipeline)
    .pipe(remove_outliers, id_list = filter_outliers(orders, orderlines))
)
```

```python
orderlines_clean = (
orderlines
    .pipe(start_pipeline)
    .pipe(remove_outliers, id_list = filter_outliers(orders, orderlines))
)
```

```python
import re

regexp_dict = {
    'Monitors':'(monitor|dell)',
    'Keyboards,pens and mice':'(keyboard|keypad|mouse|stylus|pencil|spare ball|pointer|^pen)',
    'Storage, memory cards, and drives':'(ssd|hdd|transcend|Western Digital|toshiba|drobo|harddrive|hard disk|wd red|mypassport|disk|drive|lacie|hard drive|kingston|memory|owc)',
    'Mac':'(^.{0,7}apple mac|macbook|mac mini|mac pro|imac)',
    'iPhone':'iPhone',
    'Tablets':'(iPad|wacom|tablet)',
    'iPod':'(ipod)',
    'Tools, repair, support':'(tools|screwdriver|installation kit|installation service|ifixit|repair|support)',
    'Cases and Protection':'(case|sleeve|tucano|bag|larobe|pleated|la robe|backpack|cover|housing|protect|protector|tempered glass|shell|clearguard)',
    'Watches':'(watch|smartwatch|smart watch)',
    'Surveillance, tracker, camera':'(camera|surveillance|trackr|locator|gopro|webcam)',
    'Straps':'(strap|armband|belt|bracelet|bracelets|correa)',
    'Cables and Adapters':'(cable|lightning|adapter|displayport to|connector)',
    'Refurbished':'(refurbished|reconditioned|like new|second hand)',
    'NAS Server body and packs':'(raid|synology|qnap|owc box|nas|wd my cloud|my cloud|server)',
    'Networking hardware ':'(hub|switch|router|extender|d-link|repeater|dlan)',
    'External peripherals':'Thunderbolt',
    'Battery, charging and power supply':'(battery|powerstation|charger|lampcharge|charging|powercube|power bank|power line)',
    'Drones, Robots, Sensors':'(drone|robot|sensor|climate|thermostat)',
    'Stands':'(hirise|bookarc|stand|stands|docking|dock)',
    'Headsets and Speakers':'(headphones|earphones|headset|subwoofer|microphone|speaker|airpods|powerbeats|beats)'
    
}

temp = products.copy().assign(category = 'unknown')

import numpy as np

for label, value in regexp_dict.items(): 
    regexp = re.compile(value, flags=re.IGNORECASE)    
    temp = (
    temp
        .assign(
            category = lambda x: np.where(
                (x['desc'].str.contains(regexp, regex=True)), label, x['category'])))

temp['price_cat']=pd.cut(temp['price'], bins=[0,1000,9000], labels=['cheap', 'expensive'])
    
#remove all storage items that are wrongly put into mac category
temp.loc[(temp['category'] == 'Mac') & (temp['price_cat'] == 'cheap') & (temp['desc'].str.contains(r'expansion|kit|Enlargement|NewerTech|Memory|JetDrive|OWC')), 'category'] = 'Storage, memory cards, and drives'
```
